```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv')
```

Cleansing

```python
dataset.isnull().sum()
```

```python
len(dataset)
```

```python
dataset2 = dataset[~(dataset['start_latitude'].isna() & dataset['start_longitude'].isna() & dataset['end_latitude'].isna() & dataset['end_longitude'].isna())]

#removes all the rows where given column values are null
```

```python
dataset2.head(10)
```

```python

dataset2.columns
```

```python
dataset2['total_halt_time'] = np.where(dataset2['total_halt_time'] > 60,0,dataset2['total_halt_time'])
len(dataset2['summary_for_date'].unique())
#to remove all the outliers(where the total halt time >60)
```

Data Preparation

```python
dataset2.sort_values(by=['vin','summary_for_date','hour_slot'],inplace=True) #sort dataset based on vin and hour_slot and group by vin
dataset2.reset_index(drop=True,inplace=True)
```

```python
#dataset2.loc[dataset2['vin']== 'XYZD0HEJZ714522']
```

```python
dataset2['is_bs4'] = np.where(dataset2['is_bs6']!=1,1,0)
```

```python
ECOMETV = dataset2['nomenclature_desc'].str.contains('ECOMET')
dataset2['ECOMETV'] = np.where(ECOMETV==True,1,0)
```

```python
V1612 = dataset2['nomenclature_desc'].str.contains('1612')
dataset2['1612V'] = np.where(V1612==True,1,0)
```

```python
V1618 = dataset2['nomenclature_desc'].str.contains('1618')
dataset2['1618V'] = np.where(V1618==True,1,0)
```

```python
Unknown = dataset2['nomenclature_desc'].str.contains('Unknown')
dataset2['UnknownV'] = np.where(Unknown==True,1,0)
```

```python
print(ECOMETV.unique())
```

```python
features = ['vin','hour_slot','summary_for_date','total_halt_time','start_latitude','start_longitude','end_latitude','end_longitude','is_bs4','is_bs6','ECOMETV','1612V','1618V','UnknownV','nomenclature_desc']

dataset3 = dataset2[features]
dataset3.reset_index(drop=True)
```

```python
dataset3['halt_flag'] = np.where(dataset3['total_halt_time'] == 60,1,0)

#new column halt flag, if total_halt_time =60 then flag is 1 else it is 0
```

```python
dataset3['day'] = np.where(((dataset3['hour_slot'] >= 8) & (dataset3['hour_slot'] <= 22)),1,0)
dataset3['night'] = np.where(~((dataset3['hour_slot'] >=8) & (dataset3['hour_slot'] <= 22)),1,0)

dataset3.head(10)
```

```python
dataset3['halt_flag_shift'] = dataset3.groupby('vin')['halt_flag'].shift(1)

```

```python
dataset3.loc[dataset3['vin'] == 'XYZD0JADJ086622']
```

```python
dataset3['idle_flag'] = np.where((dataset3['halt_flag'] == 1  ) & (dataset3['halt_flag_shift'] == 0 | dataset3['halt_flag_shift'].isna()),1,0)
```

```python
dataset3['cum_sum'] = dataset3.groupby(['vin'])['idle_flag'].cumsum()
```

```python
dataset_filter = dataset3.loc[dataset3['halt_flag']==1]
```

```python
#dataset_filter['sum_halt_time'] = dataset_filter.groupby(['vin','cum_sum'])[['total_halt_time']].agg('sum')
dataset_new = dataset_filter.groupby(['vin','cum_sum'])['total_halt_time'].sum().reset_index()
dataset_newtest = dataset_filter.groupby(['vin','cum_sum'])['day','night','is_bs4','is_bs6','1612V','1618V','ECOMETV','UnknownV','nomenclature_desc'].first().reset_index()
######merge based on vin and cumsum
```

```python
dataset_new=dataset_new.merge(dataset_newtest,on=['vin','cum_sum'])
```

```python
dataset_new.head(50)
```

```python
#dataset_filter.loc[dataset_filter['vin']=='XYZD0JAAJ381322']
```

```python
dataset_new1 = dataset_filter.groupby(['vin','cum_sum'])['start_latitude'].first().reset_index()
```

```python
dataset_new2 = dataset_filter.groupby(['vin','cum_sum'])['start_longitude'].first().reset_index()
```

```python
#day/night........
```

```python
dataset_final=dataset_new.merge(dataset_new1,on=['vin','cum_sum']).merge(dataset_new2,on=['vin','cum_sum'])
```

```python
dataset_final = dataset_final.loc[dataset_final['total_halt_time']>240]
```

```python
values =  dataset_final.iloc[ : ,12:14].values  #converts into a np.array
```

```python
from sklearn.cluster import DBSCAN
epsilon = 3/6371.0088

model = DBSCAN(eps=epsilon,min_samples=5,metric='haversine').fit(np.radians(values))
```

```python
cluster_labels = model.labels_
```

```python
pd.DataFrame(cluster_labels).value_counts()
```

```python
print(cluster_labels)
```

```python
cluster_dataset = dataset_final.loc[cluster_labels !=-1]
```
